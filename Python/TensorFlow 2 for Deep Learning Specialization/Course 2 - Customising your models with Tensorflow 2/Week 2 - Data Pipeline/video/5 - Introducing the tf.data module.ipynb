```python
import tensorflow as tf
```

```python
dataset = tf.data.Dataset.from_tensor_slices([1, 2, 3, 4, 5, 6])

print(dataset)
```

```python
dataset = tf.data.Dataset.from_tensor_slices([[1, 2], [3, 4], [5, 6]])

print(dataset)
```

```python
print(dataset.element_spec)
```

```python
for elem in dataset:
    print(elem)
```

```python
for elem in dataset:
    print(elem.numpy())
```

```python
dataset = tf.data.Dataset.from_tensor_slices(tf.random.uniform([128, 5]))

print(dataset)

print(dataset.element_spec)

for elem in dataset:
    print(elem.numpy())
    
```

```python
dataset = tf.data.Dataset.from_tensor_slices(tf.random.uniform([2, 4, 4, 3]))

print(dataset)

print(dataset.element_spec)

for elem in dataset:
    print(elem.numpy())
```

```python
dataset = tf.data.Dataset.from_tensor_slices(
    (
        tf.random.uniform([256, 4], minval=1, maxval=10, dtype=tf.int32),
        tf.random.normal([256])
    )
)

print(dataset.element_spec)
```

```python
dataset = tf.data.Dataset.from_tensor_slices(
    (
        tf.random.uniform([2, 4, 4, 3], minval=0, maxval=256, dtype=tf.int32),
        tf.random.normal([2])
    )
)

#print(dataset)

#print(dataset.element_spec)

for elem in dataset:
    print(elem)
```

```python
dataset = tf.data.Dataset.from_tensor_slices(
    (
        tf.random.uniform([2, 4, 4, 3], minval=0, maxval=256, dtype=tf.int32),
        tf.random.normal([2])
    )
)

#print(dataset)

#print(dataset.element_spec)

for elem in dataset.take(1):
    print(elem)
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))

print(dataset.element_spec)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

img_datagen = ImageDataGenerator(width_shift_range=0.2, horizontal_flip=True)
dataset = tf.data.Dataset.from_generator(img_datagen.flow, 
                                         args=[x_train, y_train],
                                         output_types=(tf.float32, tf.int32),
                                         output_shapes=([32, 32, 32, 3], [32, 1]))
```
